This Act Report is part of the requirements of Project 2 of the ALX-T Data Analyst Nanodegree program on Udacity.

During the exploratory analysis stage, I posed five questions to guide my analysis, leading me to the following insights:

1. There's a very positive correlation (r=0.93) between retweets and favourite count of a tweet.
2. Tweets of the doggo-puppo category overwhelmingly outperformed the rest in retweets and likes.
3. Tweeting from a web browser seems to have gathered more retweets on average.

## INSIGHTS

1. The scatterplot below shows a positive linear correlation between the number of favourites (likes) a tweet got and the number of retweets. A further examination using `.corr()` revealed that a Pearson's coefficient of (r=0.93) between retweets and favourite count of a tweet.

> So, it appears that the more retweets a WeRateDogs tweet gets, the more favourites (likes) it is likely to also have. This does not, of course, imply that the retweets cause the favourites or vice-versa. 


2. As visible from the bar charts below, tweets of the doggo-puppo category overwhelmingly outperformed the rest in retweets and favourites (likes). WeRateDogs might, thus, explore rating more dogs of the doggo-puppo category for better engagement.

3. Tweeting from a web browser seems to have garnered more retweets on average, while tweets from an iPhone garnered more favourites on average. 

> Seeing as tweets made from TweetDeck performed the worst in retweets and favourites (likes), WeRateDogs might look into tweeting less from TweetDeck in favour of tweeting more from iPhone and web browser.

## Limitations

1. 84% (1561/1851) of the dogs didn't have a category. This means conclusions about dog stae were made on a very small portion of the observations.

2. A tweet source "Vine - Make a Scene" wasn't part of the final analysis. I suspect this tweet source was dropped when I dropped retweet rows and tweets with nonstandard dog names. 

3. I couldn't query additional data on 29 tweets. 28 of the tweets threw up "No status found with that ID" error, while one threw up "Sorry, you are not authorized to see this status" error.
